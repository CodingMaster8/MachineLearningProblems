```python
#Importing Necessary Packages
import numpy as np
import skfuzzy as fuzz
import matplotlib.pyplot as plt
%matplotlib inline
#Defining the Fuzzy Range from a speed of 30 to 90
x = np.arange(30, 80, 0.1)
#Defining the triangular membership functions
slow = fuzz.trimf(x, [30, 30, 50])
medium = fuzz.trimf(x, [30, 50, 70])
medium_fast = fuzz.trimf(x, [50, 60, 80])
full_speed = fuzz.trimf(x, [60, 80, 80])
#Plotting the Membership Functions Defined
plt.figure()
plt.plot(x, full_speed, 'b', linewidth=1.5, label='Full Speed')
plt.plot(x, medium_fast, 'k', linewidth=1.5, label='Medium Fast')
plt.plot(x, medium, 'm', linewidth=1.5, label='Medium Powered')
plt.plot(x, slow, 'r', linewidth=1.5, label='Slow')
plt.title('Penalty Kick Fuzzy')
plt.ylabel('Membership')
plt.xlabel("Speed (Miles Per Hour)")
plt.legend(loc='center right', bbox_to_anchor=(1.25, 0.5),
ncol=1, fancybox=True, shadow=True)
```

#### Now, modify the functions in such a way that
##### • Medium Powered goes from 40, 50, 60
##### • Medium Fast goes from 50, 60, 70
##### Determine for both cases (original sets and modified ones) the (approx.) values for speed: 45 and 65.


```python
# Modified membership functions
medium_mod = fuzz.trimf(x, [40, 50, 60])
medium_fast_mod = fuzz.trimf(x, [50, 60, 70])

# Speeds to evaluate
speeds = [45, 65]

# Calculate memberships for original functions
for speed in speeds:
    print(f"\n--- Speed: {speed} ---")
    print("Original:")
    print(f"  Slow: {fuzz.interp_membership(x, slow, speed):.2f}")
    print(f"  Medium: {fuzz.interp_membership(x, medium, speed):.2f}")
    print(f"  Medium Fast: {fuzz.interp_membership(x, medium_fast, speed):.2f}")
    print(f"  Full Speed: {fuzz.interp_membership(x, full_speed, speed):.2f}")

    print("Modified:")
    print(f"  Medium Powered (40,50,60): {fuzz.interp_membership(x, medium_mod, speed):.2f}")
    print(f"  Medium Fast (50,60,70): {fuzz.interp_membership(x, medium_fast_mod, speed):.2f}")
```

```python
# Define the fuzzy range
x = np.arange(20, 91, 0.1)

# Updated trapezoidal membership functions
slow = fuzz.trapmf(x, [20, 30, 35, 45])
medium = fuzz.trapmf(x, [40, 50, 55, 65])
medium_fast = fuzz.trapmf(x, [60, 65, 70, 75])
full_speed = fuzz.trapmf(x, [70, 80, 85, 90])

# Plotting the membership functions
plt.figure(figsize=(10, 5))
plt.plot(x, slow, 'r', linewidth=1.5, label='Slow')
plt.plot(x, medium, 'm', linewidth=1.5, label='Medium Powered')
plt.plot(x, medium_fast, 'k', linewidth=1.5, label='Medium Fast')
plt.plot(x, full_speed, 'b', linewidth=1.5, label='Full Speed')
plt.title('Penalty Kick Fuzzy (Trapezoidal Functions)')
plt.ylabel('Membership')
plt.xlabel("Speed (Miles Per Hour)")
plt.grid(True)
plt.legend(loc='center right', bbox_to_anchor=(1.25, 0.5),
           ncol=1, fancybox=True, shadow=True)
plt.show()
```
